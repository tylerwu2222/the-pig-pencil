```python
import requests
import pandas as pd
```

```python
user_id = 'tylorherro'
league_id = '883800627342376960'
```

```python
# get rosters
response = requests.get("https://api.sleeper.app/v1/league/" + league_id + '/rosters')
rosters = response.json()
```

```python
def get_object_values(row, indices):
    values = list(row.values())
    indexed_values = [values[i] for i in indices]
    return indexed_values

def get_object_keys(row, indices):
    keys = list(row.keys())
    indexed_keys = [keys[i] for i in indices]
    return indexed_keys
```

```python
roster_df = pd.DataFrame(rosters).reset_index(drop=True)
# split settings col
roster_df[get_object_keys(roster_df.settings.iloc[0],[0,1,3,4,6,7,10,11])] = pd.DataFrame(roster_df['settings'].apply(lambda row: get_object_values(row,[0,1,3,4,6,7,10,11])).tolist())
# split metadata col
roster_df['streak'] = [i.get('streak') for i in roster_df.metadata]
roster_df['record'] = [i.get('record') for i in roster_df.metadata]
roster_df = roster_df.drop(columns=['taxi','settings','roster_id','player_map','metadata','league_id'])
roster_df
```
