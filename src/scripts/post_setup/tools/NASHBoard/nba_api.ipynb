```python
from nba_api.stats.endpoints import commonallplayers
from nba_api.stats.endpoints import commonteamroster
from nba_api.stats.static import players
from nba_api.stats.static import teams

import json
import requests
import os
import time

from bs4 import BeautifulSoup
```

```python
AP_dict = commonallplayers.CommonAllPlayers().get_dict()['resultSets'][0]['rowSet']
AP_dict
```

```python
data_folder = '../../components/Tools/NASHBoard/Data/'
```

```python
# store all active players in json
active_players = players.get_active_players()
active_players
with open("active_players.json", "w") as outfile:
    json.dump(active_players, outfile, indent=4)
```

```python
# get all player profile pics from nba's CDN by id and save to folder
root = 'https://cdn.nba.com/headshots/nba/latest/1040x760/'
player_ids = [p.get('id') for p in active_players]
for p_id in player_ids:
    src = root + str(p_id) + '.png'
    img_data = requests.get(src).content
    with open(os.path.join('player_headshots', str(p_id) + '.jpg'), 'wb') as handler:
        handler.write(img_data)
```

```python
# get team logos and save to folder
url = 'https://www.sportslogos.net/teams/list_by_year/62022/2022_NBA_Logos/'
page = requests.get(url)
soup = BeautifulSoup(page.content, 'html.parser')
a_tags = soup.find("div",{"id":"team"}).find_all("a")
img_urls = ['https://www.sportslogos.net' + a['href'] for a in a_tags]
img_urls = img_urls[:len(img_urls)-1]
img_urls
```

```python
img_cdns = []
for url in img_urls:
    print(url)
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser') 
    img_cdn = soup.find("div",{"id":"mainLogo"}).find_all('img')[0]['src']
    img_cdns.append(img_cdn)
img_cdns
```

```python
f = open('teams.json')
team_names = [t['teamName'] for t in json.load(f)]
team_names
```

```python
# get team rosters and add to json
f = open('../teams.json')
team_dict = json.load(f)
team_ids = [t['teamId'] for t in team_dict]
for i,id in enumerate(team_ids):
    print(i)
    time.sleep(2)
    roster = commonteamroster.CommonTeamRoster(team_id = id).get_dict()['resultSets'][0]['rowSet']
    roster_list = []
    for r in roster:
        player = {}
        player['id'] = r[-2]
        player['player'] = r[3]
        # print(header,'\n',roster)
        roster_list.append(player)
    team_dict[i]['roster'] = roster_list
team_dict
```

```python
with open(data_folder + "/teams.json", "w") as outfile:
    json.dump(team_dict, outfile)
```

```python
for i, logo in enumerate(img_cdns):
    # print(i)
    img_data = requests.get(logo).content
    with open(os.path.join('team_logos', team_names[i]  + '.jpg'), 'wb') as handler:
        handler.write(img_data)
```
