```python
import pandas as pd
from nba_api.stats.static import players
from nba_api.stats.endpoints import playergamelog, commonplayerinfo,playercareerstats
```

```python
players_dict = {player['full_name']:player['id'] for player in players.get_players()}
```

```python
# filter for players who were drafted since 1970 and also in first 2 rounds
drafted_players = pd.read_csv('nba_drafts.csv')
```

```python
top_round_players = list(drafted_players.loc[drafted_players.Round <=2,].Player)
top_round_players
```

```python
# all_columns = playercareerstats.PlayerCareerStats(player_id=2544).get_data_frames()[0].columns
# all_columns
select_columns = ['PLAYER_ID','SEASON_ID','LEAGUE_ID','TEAM_ID',
                  'PLAYER_AGE','GP','GS','MIN']
```

```python
top_players_dict.values()
```

```python
len(top_players_dict)
```

```python
top_players_dict = {player_name:player_id for (player_name, player_id) in players_dict.items() if player_name in top_round_players}
top_players_dict
```

```python
# get player info for each player
player_info_list = []
i = 0
for p_name, p_id in top_players_dict.items():
    i += 1
    if i%5==0:
        print(i, '/2015')
    player_info = playercareerstats.PlayerCareerStats(
        player_id=p_id).get_data_frames()[0]
    # when TEAM_ID = 0 = total for season
    player_info = player_info.loc[:, select_columns]
    player_info['PLAYER_NAME'] = p_name
    player_info_list.append(player_info)
player_info_list[0]
```

```python
combined_table_df = pd.concat(player_info_list, axis=0, ignore_index=True)
combined_table_df
```

```python
# export to csv
combined_table_df.to_csv('player_seasons.csv', index=False)
```
