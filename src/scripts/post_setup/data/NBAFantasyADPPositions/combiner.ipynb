```python
import pandas as pd
```

```python
# read all data and combine into one excel
# years = list(range(21,24))
years = list(range(21,23)) # 21 and 22
dataframes = []
for year in years:
    data = pd.read_excel('nba-adp-All-' + str(year) + '.xlsx')
    data['AVG'] = pd.to_numeric(data['AVG'], errors='coerce')
    data = data.dropna(subset=['AVG'])
    data.rename( columns={'Unnamed: 0':'Rank'}, inplace=True )
    data['Year'] = year
    data = data.sort_values(by=['AVG'])
    dataframes.append(data)
all_data = pd.concat(dataframes, axis=0, ignore_index=True)
all_data.head()
```

```python
all_data['Rank'] = all_data['Rank'].astype(str).str.replace('T', '', regex=False)
all_data['Rank'].value_counts()
```

```python
all_data.to_csv('nba-adp-21-22.csv',index=False)
```
