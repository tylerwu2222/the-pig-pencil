```python
import pandas as pd
import string
```

```python
usernames = ['gnome','andyvo','speckled mug','jack-o-pie','megh']
codes = list(string.ascii_lowercase)[:5] 
```

```python
data = pd.read_csv('mini_scores_wide.csv')
data = data.rename(columns=dict(zip(usernames,codes))) # code users
data.head()
```

```python
def get_expanding(col,stat='sum'):
    return getattr(col.expanding(1), stat)()
```

```python
# get running min, max, avg, total...
stats = ['sum','mean','median','min','max']
for s in stats:
    for c in codes:
        data[c + '_' + s] = get_expanding(data[c],s)
data.tail()
```

```python
data.to_csv('mini_score_c_wide.csv',index=False)
```
