```python
import json
import pandas as pd
```

```python
json_file = '../../../src/components/BlogPosts/BerkeleyNature/Data/image_data_complete.json'
csv_file = '../../../src/components/BlogPosts/BerkeleyNature/Data/image_metadata.csv'
```

```python
csv = pd.read_csv(csv_file)
# csv
```

```python
with open(json_file,'r') as f:
    json_csv = pd.DataFrame(json.loads(f.read()))
# json_csv
```

```python
csv = csv.sort_values(by=['img'])
json_csv = json_csv.sort_values(by=['img'])
# json
all(json_csv['img'].unique() == csv['img'].unique())
```

```python
# merge csv and json
# joined_df = pd.merge(json_csv, csv, on="img", how="outer")
joined_df = csv
joined_df['tags'] = json_csv['tags']
joined_df.index = joined_df['img']
joined_df
```

```python
output_file = '../../../src/components/BlogPosts/BerkeleyNature/Data/image_metadata.json'
joined_df.to_json(output_file, orient='index')
```
